```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit = pd.read_csv("customer-credit.csv")

credit.head()
```

```python
plt.scatter(data=credit, x="Income", y="Balance")
```

```python
import statsmodels.formula.api as smf

result = smf.ols(formula="Balance ~ Income", data=credit).fit()

result.summary()
```

Balance = 246.5148 + 6.0484 * Income

```python
result.params[0]
```

```python
plt.scatter(data=credit, x="Income", y="Balance")

x = np.arange(0, 180, 10)
y = result.params[0] + result.params[1] * x

plt.plot(x, y)
```
