# Python for plotting (data visualization)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# line plots

x = np.arange(start=0, stop=10, step=0.5)

plt.plot(x, np.sin(x), marker='o', color='red', alpha=0.6, 
         linestyle='dotted', lineWidth=5, markersize=10, label='sin(x)')

plt.plot(x, np.cos(x), marker='o', color='blue', alpha=0.6, 
         linestyle='dotted', lineWidth=5, markersize=10, label='cos(x)')

plt.legend()
```

```python
credit = pd.read_csv("customer-credit.csv") 

plt.hist(credit["Income"], bins=np.arange(0, 200, 10), edgeColor='black', faceColor='green')

plt.xticks(np.arange(0, 200, 20))

plt.show()
```
