```python
# open a csv file

import pandas as pd

# read the csv file
crash = pd.read_csv('ped-crashes.csv')

crash.head()
```

```python
crash.columns = ['year', 'month', 'day', 'time_of_day', 'day_of_week', 'city', 'intersection', 'hit_and_run', 'lighting', 'weather', 'speed_limit', 'worst_injury', 'party_type', 'age', 'gender']

# Find the percentage of crashes that are fatal
value = round(crash['worst_injury'].value_counts(normalize=True)*100, 2)
print(value)

# Find the percentage of crashes that are fatal
value = crash.query('worst_injury == "Fatal injury (K)"').shape[0] / len(crash) * 100
print(str(round(value, 2))+"%")
```
